Penjelasan nama kolom :


*   PassengerId : nomor indentifikasi UNIK
*   Pclass : kelas tiket penumpang (1 : kelas tertinggi, 2 kelas reguler, 3 kelas terendah dalam kapal)
*   SibSp : jumlah saudara kandung atau pasangan yang ikut dalam kapal
*   Parch (Parents/Children) : jumlah orang tua atau anak yang ikut dalam kapal
*   Fare : harga tiket
*   Cabin : nomor cabin
*   Embarked : pelabuhan tempat penumpang naik ke kapal











```python
x = df['Age'].mean()
print(x)
```

```python
df1 = df
df1["Age"].fillna(30.0, inplace = True) #untuk mengganti nilai NaN di Age dengan hasil rata rata yaitu 30
df1
```

```python
df1["Ticket"] = df["Ticket"].str.replace(" ", "")  # Hapus spasi pada kolom ticket
df1
```

```python
#menghapus tanda baca / dan . pada kolom ticket
df1["Ticket"] = df1["Ticket"].str.replace(r"[./]", "", regex=True)
df1
```

```python
#menghapus tulisan huruf pada kolom tiket
df1["Ticket"] = df1["Ticket"].str.replace(r"\D", "", regex=True)
df1
```

```python
#menghapus data yang sama pada tabel
df1.drop_duplicates(inplace=True)
df1
```

```python
print(df.duplicated())
```

```python
df1['Fare'] = df1['Fare'].round(1)  # Membulatkan ke 1 desimal
df1
```

```python
#mengisi kolom Cabin yang NaN diganti unknown
df1['Cabin'].fillna('Unknown', inplace=True)
df1
```

```python
#mengubah tipe data ke string untuk mengubah kolom tiket
df1['Ticket'] = df1['Ticket'].astype(str)
```

```python
#mengubah format agar memikiki 7 angka (untuk yang kurang ditambah 0 pafda angka sebelumya)
df1['Ticket'] = df1['Ticket'].str.zfill(7)
df1
```

```python
#mengubah isi kolom embarked ke nama pelabuhan penumpang naik ke kapal
df1['Embarked'] = df1['Embarked'].replace({'C': 'Cherbourg', 'Q': 'Queenstown', 'S': 'Southampton'})
df1
```

```python
df1.to_csv('cleandata.csv', index=False)
df1
```
